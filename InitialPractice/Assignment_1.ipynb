Question 1. The first question involves arithmetic with a small application of loops. You
are given an array of n integers, that is, a1, a2, . . . , an, and you are required to output the
maximum and minimum among them. Hint: Try with n=3, then generalise your arguement.

```python
# Question 1

print("Enter an array of numbers for which you want to find Max and Min: ")
print("Numbers should be separated by commas")
nos = input()
nos1 = nos.split(',')
nos2 = [int(x) for x in nos1]

a = nos2[0]
b = nos2[1]
if a >= b:
  max = a
  min = b
else:
  max = b
  min = a

for i in range (2, len(nos1)):
  if nos2[i] >= max:
    max = nos2[i]

for i in range (2, len(nos1)):
  if nos2[i] <= min:
    min = nos2[i]

print("The maximum number is", max)
print("The minimum number is", min)
```

Question 2. This question is for familiarity with strings. You are needed to input a string
and check if it is a palindrome. Hint: Read a bit about two pointer method.

```python
# Question 2

hi = input("Enter string: ")
str = hi.lower()
left = 0
right = len(str)-1

while left < right:
  if str[left]==str[right]:
    left += 1
    right -= 1
  else:
    print("The given string is not a palindrom")
    break
else:
  print("The given string is a palindrome")
```

Question 3. You will need a basic understanding of sorting for this question. Note that
there is a direct python function to sort a sequence of integers. You are given an array of n
integers, that is, a1, a2, . . . , an, and a number x. You need to find a pair of elements whose
absolute difference is x. You are assured that such a pair exists in the array. Hint: What if
you sort the array.

```python
# Question 3

print("Enter an array of numbers ")
print("Numbers should be separated by commas")
nos = input()
diff = int(input("Enter difference: "))
nos1 = nos.split(',')
nos2 = [int(x) for x in nos1]
nos2.sort()
print(nos2)
left = 0
right = 1
found = False
while right < len(nos2):
  if nos2[right]-nos2[left] > diff:
    left += 1
  elif nos2[right]-nos2[left] < diff:
    right += 1
  else:
    print("The numbers whose difference is", diff, "is", nos2[left], "and", nos2[right])
    left += 1
    right += 1
    found = True

if found == False:
  print("Pair of numbers whose difference is", diff, "does not exist.")
```
